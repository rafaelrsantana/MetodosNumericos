**Primeira Questão**

```python
print("PRIMEIRA QUESTÃO")
print("\n")

def conversor(binary):
    sinal, expoente, mantissa = binary.split(' ')
    s = -1 if sinal == '1' else 1
    expo = sum(2 ** (11 - (e + 1)) for e, bit in enumerate(expoente) if bit == '1')
    mant = sum(1 / (2 ** (m + 1)) for m, bit in enumerate(mantissa) if bit == '1')
    decimal = s * (1 + mant) * (2 ** (expo - 1023))
    return decimal

test_cases = [
    ("0 10000001010 1001001100000000000000001000000000001000000001000000", "A"),
    ("1 10000001010 1011001100000000001000000000000010000000000000000000", "B"),
    ("0 01101011111 0001001100000000000000100000000000000000000001000000", "C"),
    ("1 01011111011 0101001100000000000000000000000001000000001000000001", "D")
]

for binary, letter in test_cases:
    print(f"Letra {letter}")
    print("\n")
    print("Binário:", binary)
    print("Decimal:", conversor(binary))
    print("\n")
```

**Segunda Questão**

```python
def error(p, p2):
    error_absol = abs(p - p2)
    error_relat = error_absol / abs(p)
    print("Erro absoluto:", error_absol)
    print("Erro relativo:", error_relat)

print("SEGUNDA QUESTÃO")
print("\n")

test_cases = [
    (1, 0.9994, "A"),
    (124, 7, "B"),
    (22026.465794806716516957900645284, 22000, "C"),
    (40320, 39900, "D")
]

for p, p2, letter in test_cases:
    print(f"Letra {letter}")
    print("\n")
    error(p, p2)
    print("\n")
```

**Terceira Questão**

```python
def intervalo(p):
    error = p * 0.01
    lower_bound = p - error
    upper_bound = p + error
    print("p* ≥ ", upper_bound)
    print("p* ≤ ", lower_bound)

print("TERCEIRA QUESTÃO")
print("\n")

test_cases = [150, 900, 1500, 90]
letters = ['A', 'B', 'C', 'D']

for p, letter in zip(test_cases, letters):
    print(f"Letra {letter}")
    print("\n")
    intervalo(p)
    print("\n")
```

```python
import math

def truncamento(num, digits):
    multiplier = 10 ** digits
    num_truncado = int(num * multiplier) / multiplier
    return num_truncado

def arredondamento(num, digits):
    multiplier = 10 ** digits
    num_arredondado = round(num, digits)
    return num_arredondado

def erro_absoluto(num, num_aprox):
    e_abs = abs(num - num_aprox)
    return e_abs

def erro_relativo(num, num_aprox):
    e_abs = abs(num - num_aprox)
    e_rel = e_abs / abs(num)
    return e_rel

def calcular_letra_a():
    num = 133 - 0.499
    return num

def calcular_letra_b():
    a = 1 / 3
    b = -123 / 4
    c = 1 / 6
    delta = b ** 2 - 4 * a * c
    x1 = (-b + math.sqrt(delta)) / (2 * a)
    x2 = (-b - math.sqrt(delta))/(2*a)

    return x1

def calcular_letra_c():
    e = math.e
    num = ((13 / 14) - (6 / 7)) / (2 * e - 5.4)
    return num

# Execução do programa

print("QUARTA QUESTÃO")
print("\n")

letras = ['A', 'B', 'C']
funcs = [calcular_letra_a, calcular_letra_b, calcular_letra_c]

for letter, function in zip(letras, funcs):
    print(f"Letra {letter}:")
    print("\n")
    num = function()
    print(f"Resultado original: {num}")

    for digits in [3, 4]:
        print(f"Truncamento de {digits} dígitos:")
        num_truncado = truncamento(num, digits)
        e_abs_trunc = erro_absoluto(num, num_truncado)
        e_rel_trunc = erro_relativo(num, num_truncado)
        print(f"Num truncado: {num_truncado}")
        print(f"Erro absoluto: {e_abs_trunc}")
        print(f"Erro relativo: {e_rel_trunc}")
        print("\n")

        print(f"Arredondamento de {digits} dígitos:")
        num_arredondado = arredondamento(num, digits)
        e_abs_arred = erro_absoluto(num, num_arredondado)
        e_rel_arred = erro_relativo(num, num_arredondado)
        print(f"Num arredondado: {num_arredondado}")
        print(f"Erro absoluto: {e_abs_arred}")
        print(f"Erro relativo: {e_rel_arred}")
        print("\n")
```

```python
print("SEXTA QUESTÃO")

def funcao(x):
    return x ** 3 - 4 * (x ** 2) - 10


def derivada(x):
    return 3 * (x ** 2) + 8 * x

def newton(funcao, derivada, p0, tolerancia, limite):
    print("\nMétodo de Newton\n")

    for i in range(1, limite + 1):
        p = p0 - funcao(p0) / derivada(p0)

        if abs(p - p0) < tolerancia:
            print("Operação concluída na", i, "º iteração.")
            print("Resultado:", p)
            return

        p0 = p

    print("Falha encontrada após", limite, "iterações")



def bisseccao(funcao, a, b, tolerancia, limite):
    print("Método da Bissecção\n")

    for i in range(1, limite + 1):
        p = a + (b - a) / 2
        fP = funcao(p)

        if fP == 0 or abs(b - a) < tolerancia:
            print("Operação concluída na", i, "º iteração.")
            print("Resultado:", p)
            return

        if funcao(a) * fP > 0:
            a = p
        else:
            b = p

    print("Falha encontrada após", limite, "iterações")




p0 = 2
pA = 0
pB = 1
tolerancia = 1e-20
max_iterac = 1000

bisseccao(funcao, pA, pB, tolerancia, max_iterac)
newton(funcao, derivada, p0, tolerancia, max_iterac)
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors

def funcao_a(x):
    return x - (2**(-x))

def points(Xs,Ys, func):
    y=0
    Xs.sort(reverse=False)
    for x in Xs:
      y = func(x)
      Ys.append(y)

def derivada_a(x):
    return ((math.log(2))/2**x) + 1

def funcao_b(x):
    return x + 1 - 2*math.sin(math.pi*x)

def derivada_b(x):
    return 1 - 2 * math.pi * math.cos(math.pi*x)

def funcao_c(x):
    return 2 * x * math.cos(2 * x) - (x + 1)**2

def derivada_c(x):
    return 2 * math.cos(2*x) - 4*x * math.sin(2 * x) - 2*(x + 1)

def funcao_d(x):
    return math.log(x) - 2**x + x**2

def derivada_d(x):
    return - math.log(2) * 2**x + 2 * x + 1/x

tolerancia = 1e-10
itiMax = 1000

def graphic(XList, YList):
    n = len(XList)
    cmap = plt.get_cmap('cividis')
    norm = mcolors.Normalize(vmin=0, vmax=n-1)

    plt.plot(XList, YList, label='Curva Poligonal', color=(0.5, 0.5, 1.0, 0.8))
    plt.scatter(XList, YList, c=np.arange(n), cmap=cmap, norm=norm, label='Pontos', edgecolors='black')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('Gráfico y = f(x)')
    plt.legend()
    plt.grid(True)
    plt.colorbar(label='Índice Pontos')
    plt.show()

def newton(funcao,derivada,p0,tol,n_max):
  i = 1
  p = 0
  raizes = []
  print("\n--> Newton\n")

  while i<= n_max:
    p = p0 - (funcao(p0)/derivada(p0))
    raizes.append(p)

    if abs(p - p0) < tol:
      print("Operação feita na",i,"º iteração.")
      print("Resultado:", p)
      print("Aproximação da raiz:", raizes)
      return raizes, i
    i = i + 1
    p0 = p

  print("Falha de método após",n_max,"iterações\n")

def bisseccao(funcao,a,b,tol,n_max):
  i = 1
  p = 0
  fA = funcao(a)
  raizes = []
  print("\n--> Bissecção\n")

  while i <= n_max:
    p = a + ((b-a)/2)
    fP = funcao(p)
    raizes.append(p)

    if fP == 0 or abs(b-a) < tol:
      print("Operação feita na",i,"º iteração.")
      print("Resultado:", p)
      print("Lista de aproximações da raiz:", raizes)
      return raizes, i

    i = i + 1
    if fA * fP > 0: a = p
    else: b = p

  print("Falha de método após",n_max,"iterações\n")



print('Letra A : ')
Xs, i = newton(funcao_a, derivada_a, 0.5, tolerancia, itiMax)
Ys = []
points(Xs,Ys, funcao_a)
graphic(Xs, Ys)

Xs, i = bisseccao(funcao_a, 0, 1, tolerancia, itiMax)
Ys = []
points(Xs,Ys, funcao_a)
graphic(Xs, Ys)
```
